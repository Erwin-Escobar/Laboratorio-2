```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def mostrar_imagen_3d(imagen):
    # Convertir la imagen a escala de grises si no lo est√°
    if len(imagen.shape) == 3:
        imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
    
    # Obtener las dimensiones de la imagen
    alto, ancho = imagen.shape
    x = np.arange(0, ancho)
    y = np.arange(0, alto)
    x, y = np.meshgrid(x, y)
    
    # Crear la figura y el eje 3D
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    
    # Graficar la superficie
    ax.plot_surface(x, y, imagen, cmap='gray', edgecolor='none')
    
    # Etiquetas de los ejes
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Intensidad de Gris')
    ax.set_title('Imagen en 3D')
    
    # Mostrar la gr√°fica
    plt.show()

# Cargar la imagen en escala de grises
ruta_imagen = r"C:\Users\egesc\OneDrive - Universidad Galileo\2024\Trimestre 3\TEXT MINING & IMAGE RECOGNITION\Proyecto_2\imagen.jpg"
imagen = cv2.imread(ruta_imagen, cv2.IMREAD_GRAYSCALE)

# Mostrar la imagen en 3D
mostrar_imagen_3d(imagen)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def transformacion_negativa(imagen):
    # Transformaci√≥n negativa
    imagen_negativa = 255 - imagen
    return imagen_negativa

def transformacion_lineal(imagen, a=1.0, b=0):
    # Transformaci√≥n lineal
    imagen_lineal = cv2.convertScaleAbs(imagen, alpha=a, beta=b)
    return imagen_lineal

def transformacion_logaritmica(imagen, c=255):
    # Transformaci√≥n logar√≠tmica
    imagen_log = c * np.log1p(imagen.astype(np.float32) / 255.0)
    imagen_log = cv2.normalize(imagen_log, None, 0, 255, cv2.NORM_MINMAX)
    imagen_log = np.uint8(imagen_log)
    return imagen_log

def transformacion_exponencial(imagen, gamma=1.0):
    # Transformaci√≥n exponencial
    imagen_exp = np.power(imagen / 255.0, gamma) * 255
    imagen_exp = np.uint8(imagen_exp)
    return imagen_exp

def mostrar_imagenes(imagenes, titulos):
    plt.figure(figsize=(15, 10))
    for i, (img, titulo) in enumerate(zip(imagenes, titulos)):
        plt.subplot(2, 2, i + 1)
        plt.imshow(img, cmap='gray')
        plt.title(titulo)
        plt.axis('off')
    plt.tight_layout()
    plt.show()

# Cargar la imagen en escala de grises
ruta_imagen = r"C:\Users\egesc\OneDrive - Universidad Galileo\2024\Trimestre 3\TEXT MINING & IMAGE RECOGNITION\Proyecto_2\imagen.jpg"
imagen = cv2.imread(ruta_imagen, cv2.IMREAD_GRAYSCALE)

# Aplicar transformaciones
imagen_negativa = transformacion_negativa(imagen)
imagen_lineal = transformacion_lineal(imagen)
imagen_logaritmica = transformacion_logaritmica(imagen)
imagen_exponencial = transformacion_exponencial(imagen)

# Mostrar todas las im√°genes transformadas
imagenes = [imagen_negativa, imagen_lineal, imagen_logaritmica, imagen_exponencial]
titulos = ['Negativa', 'Lineal', 'Logar√≠tmica', 'Exponencial']

mostrar_imagenes(imagenes, titulos)
```

## Kernel Laplaciano (Laplacian Kernel)
### Descripci√≥n:
El Laplaciano es un operador de detecci√≥n de bordes que utiliza un kernel para encontrar √°reas de cambio r√°pido en la intensidad de la imagen, como los bordes. 
* Kernel: El kernel Laplaciano m√°s com√∫n es una matriz de 3x3.
### Caracter√≠sticas:
* Detecci√≥n de Bordes: Resalta los bordes al encontrar las √°reas donde hay un cambio significativo en la intensidad.
* No es Direccional: No tiene preferencia por una direcci√≥n espec√≠fica de los bordes, ya que considera los cambios en todas las direcciones.
* Ruido: Puede ser sensible al ruido, ya que el Laplaciano es una segunda derivada que amplifica el ruido en la imagen.
Aplicaci√≥n:

### Convoluci√≥n:
Se aplica el kernel Laplaciano a la imagen mediante la operaci√≥n de convoluci√≥n para obtener la imagen de salida.
### Algoritmo de Canny
### Descripci√≥n:
El algoritmo de Canny es un enfoque m√°s sofisticado y robusto para la detecci√≥n de bordes que utiliza varios pasos para detectar bordes en una imagen.

Pasos del Algoritmo de Canny:

* Filtrado: Aplica un filtro gaussiano para suavizar la imagen y reducir el ruido.
* Gradiente: Calcula el gradiente de intensidad en cada p√≠xel para encontrar las magnitudes y direcciones de los bordes.
* Supresi√≥n de No M√°ximos: Elimina los p√≠xeles que no son m√°ximos locales en la direcci√≥n del gradiente, afinando as√≠ los bordes.
* Histeresis: Utiliza dos umbrales para decidir si un p√≠xel es parte de un borde o no, con un umbral alto para detectar bordes fuertes y un umbral bajo para conectarlos.

### Caracter√≠sticas:

* Detecci√≥n de Bordes: Proporciona bordes m√°s delgados y continuos en comparaci√≥n con el Laplaciano.
* Direccionalidad: Detecta bordes en diferentes direcciones y proporciona una mejor precisi√≥n en la ubicaci√≥n de los bordes.
* Ruido: Menos sensible al ruido debido al filtrado gaussiano previo.

Comparaci√≥n:

* Precisi√≥n: El algoritmo de Canny generalmente proporciona resultados m√°s precisos y robustos que el kernel Laplaciano, ya que realiza un procesamiento m√°s complejo y utiliza t√©cnicas avanzadas como la supresi√≥n de no m√°ximos y la histeresis.
* Complejidad: Canny es m√°s complejo en t√©rminos de c√°lculo y par√°metros, mientras que el kernel Laplaciano es m√°s simple pero menos robusto.
* Sensible al Ruido: El Laplaciano puede ser m√°s sensible al ruido, mientras que Canny incluye un paso de suavizado para mitigar esto.

### Resumen:

Kernel Laplaciano: Utiliza una segunda derivada para detectar bordes, simple y directo, pero puede ser sensible al ruido.
Algoritmo de Canny: Utiliza un enfoque m√°s completo con filtrado, gradientes, supresi√≥n de no m√°ximos y histeresis, proporcionando bordes m√°s precisos y menos afectados por el ruido.
